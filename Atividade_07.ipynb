##Estudo da amostra de dados do dataset

```python
df2 = df1.copy()
```

```python
df2['sexo'].unique()
```

```python
quantidade_por_sexo = df2["sexo"].value_counts().sort_values(ascending = False).to_frame().reset_index()
quantidade_por_sexo.columns = ['Sexo','Quatidade (Pessoas)']
quantidade_por_sexo
```

```python
df2['raca'].unique()
```

```python
quantidade_por_raca = df2["raca"].value_counts().sort_values(ascending = False).to_frame().reset_index()
quantidade_por_raca.columns = ['Raça','Quatidade (Pessoas)']
quantidade_por_raca
```

```python
plt.bar(quantidade_por_raca['Raça'],quantidade_por_raca['Quatidade (Pessoas)'], edgecolor = 'black')
plt.title('Pacientes por raça', fontsize = 25)
plt.xlabel('Raça', fontsize = 20)
plt.ylabel('Quantidade (Pessoas)', fontsize = 20)
```

```python
df2['municipio_residencia'].unique()
```

```python
quantidade_por_municipio = df2["municipio_residencia"].value_counts().sort_values(ascending = False).to_frame().reset_index()
quantidade_por_municipio.columns = ['Município','Quatidade (Pessoas)']
quantidade_por_municipio
```

```python
df2['bairro'].unique()
```

```python
quantidade_por_bairro = df2["bairro"].value_counts().sort_values(ascending = False).to_frame().reset_index()
quantidade_por_bairro.columns = ['Bairro','Quatidade (Pessoas)']
quantidade_por_bairro
```

```python
df2['profissional_saude'].unique()
```

```python
profissional_de_saude = df2["profissional_saude"].value_counts().sort_values(ascending = False).to_frame().reset_index()
profissional_de_saude.columns = ['Profissional de saúde','Quatidade (Pessoas)']
profissional_de_saude
```

```python
plt.bar(profissional_de_saude['Profissional de saúde'],profissional_de_saude['Quatidade (Pessoas)'], edgecolor = 'black')
plt.xlabel('Profissional de saúde', fontsize = 20)
plt.ylabel('Quantidade (Pessoas)', fontsize = 20)
```
